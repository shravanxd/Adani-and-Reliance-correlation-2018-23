```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from a CSV file
data = pd.read_csv("stocks.csv")

# Extract the closing prices of the two stocks
reliance = data['reliance']
adani = data['adani']

# Calculate the moving averages of the two stocks
reliance_ma = reliance.rolling(window=2).mean()
adani_ma = adani.rolling(window=2).mean()

# Calculate the correlation between the two stocks
correlation = np.corrcoef(reliance, adani)[0,1]

# Plot the moving averages of the two stocks
plt.plot(reliance_ma, label="Reliance Moving Average")
plt.plot(adani_ma, label="Adani Moving Average")
plt.legend()

# Display the correlation coefficient
print("The correlation between Reliance and Adani is:", correlation)

# Show the plot
plt.show()
```
